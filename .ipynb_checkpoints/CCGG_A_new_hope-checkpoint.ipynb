# Numerical Solution

```python
class AXIS_interval:
    
    def __init__(self, x_iv):
        self.m = x_iv[2]+1
        self.x = np.linspace(x_iv[0],x_iv[1],self.m)
        self.xrev = np.linspace(x_iv[1],x_iv[0],self.m)
        self.y = np.zeros((3,self.m))
        return


class AXIS_global:
    
    def __init__(self, x_iv, x_iv1):
        self.AXI = []
        # Input of xs,xe,n_sample per interval----------
        for i in range(len(x_iv)):
            self.AXI.append(AXIS_interval(x_iv[i]))
        self.iv1 = x_iv1     # index of interval [..., 1]
        
        self.niv = len(self.AXI)    # total no. of intervals
        self.n1 = 0                 # index of x=1
        if x_iv1 >= 0:
            for i in range(self.iv1+1):
                self.n1 += self.AXI[i].m
            self.n1 -= self.iv1+1

        self.m = 0
        for i in range(self.niv):
            self.m += self.AXI[i].m
        self.m -= self.niv-1
        self.x = np.zeros(self.m)
        k = 0
        for i in range(self.niv):
            for j in range(self.AXI[i].m):
                self.x[k+j] = self.AXI[i].x[j]
            k += self.AXI[i].m-1
        return
```

```python
def prepare_solving():
    # start values, y0[2] = s(1) will be calculated
    y0 = np.zeros(3)
    y0[0] =  1    # a(1)
    y0[1] =  1    # H(1)

    # t-axis
    t_iv = []
    t_iv.append([-0.5, 0,  400])
    t_iv.append([0, 1,  200])
    t_iv.append([1, 3, 200])
    t_iv1 = 1
    AX = AXIS_global(t_iv, t_iv1)
        
    y    = np.zeros((3,AX.m))
    yp   = np.zeros(3)
    
    # z-axis
    #nz   = 200
    #zmin = redshift[0]-0.0001
    #zmax = redshift[len(redshift)-1]+0.0001
    zmin = 0
    zred = 2
    nz   = 200
    zmax = 1100
    ncmb = 2000
    z    = np.linspace(zmin, zred, nz)
    zcmb = np.linspace(zred, zmax, ncmb)
    nt   = nz+ncmb-1
    zt   = np.zeros(nt)
    for i in range(nt):
        zt[i] = z[i] if i <= nz-1 else zcmb[i-nz+1]
    #print(zt)
    
    di   = np.zeros(nz)
    Hz   = np.zeros(nt)
    mu   = np.zeros(nz)
    rcd  = np.zeros(nt)  #comoving distance
    mu_s = np.zeros(nt)

    return AX, y0, y, yp, nz, z, Hz, di, mu, zt, rcd, mu_s
```

```python
# Friedman equations
def FL_equations(x, y, yp, *args):
    Om   = args[0]
    Orad = args[1]
    Ol   = args[2]
    Ok   = args[3]
    Os   = args[5]
    Og   = args[6]
    
    err  = 0
    Os2  = Os/2
    a    = y[0]
    H    = y[1]
    s    = y[2] 

    if a > 1.e-12:
        a2 = a**2
        a3 = a**3
        H2 = H**2
        s2 = s**2
        M  = 0.25*Om/a3+Ol
        V0 = -Om/a-Orad/a2-Ol*a2
        yp[0] = a*H
        yp[1] = -2*H2+2*M+Ok/a2-(Os-1)*s2
        ks = (H2*s2-M*(3*Om/(a3*4)+Orad*a**(-4))
              +Os2*s2*(H2+(Os2-1)*s2-Ok/a2)
              +(Og-M)*(H2+V0/a2-Ok/a2+(Os-1)*s2))
        if ks < 1.e-12: err = 2
        else: yp[2] = -H*s+2*np.sqrt(ks)
    else:
        err = 1
    return yp, err
```

```python
# 1st order Runge Kutta
def FO_RK4(f, x, y, yp, *args):
    m   = len(x)         
    h   = x[1]-x[0]
    h2  = h/2
    err = 0
    nerr = -1
 
    for i in range(1, m):
        yh = y[:, i-1]
        k1, err1 = f(x[i-1],    yh,      yp, *args)
        k1 = k1*h
        k2, err2 = f(x[i-1]+h2, yh+k1/2, yp, *args)
        k2 = k2*h
        k3, err3 = f(x[i-1]+h2, yh+k2/2, yp, *args)
        k3 = k3*h
        k4, err4 = f(x[i-1]+h,  yh+k3,   yp, *args)
        k4 = k4*h
        y[:,i] = yh+(k1+2*k2+2*k3+k4)/6

        if err1+err2+err3+err4 != 0:
            err = 1
            nerr = i
            break
    return err, nerr
```

```python
#Solving the differential equations
def calc_aHs(AX, y0, yp, *args):   
    imin  = 0
    imax  = AX.m-1
    y0[2] = args[7]  # s(1)
    
    k = AX.n1
    y[:,k] = y0
    for i in range(AX.iv1,-1,-1):
        xi = AX.AXI[i].xrev
        yi = AX.AXI[i].y
        yi[:,0] = y[:,k]
        err, nerr = FO_RK4(FL_equations, xi, yi, yp, *args)
        for j in range(AX.AXI[i].m):
            y[:,k-j] = yi[:,j]
        if err != 0: 
            imin = k-nerr+2
            break
        k -= AX.AXI[i].m-1
    
    k = AX.n1
    y[:,k] = y0
    for i in range(AX.iv1+1,AX.niv):
        xi = AX.AXI[i].x
        yi = AX.AXI[i].y
        yi[:,0] = y[:,k]
        err, nerr = FO_RK4(FL_equations, xi, yi, yp, *args)
        for j in range(AX.AXI[i].m):
            y[:,k+j] = yi[:,j]
        if err != 0:
            imax = k+nerr-1
            break
        k += AX.AXI[i].m-1
    return imin, imax, AX.x, y
```

```python
def test_aHs(AX, y0, yp, *args):
    imin, imax, x, y = calc_aHs(AX, y0, yp, *args)

    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.set_xlim(-0.5, 3.)
    ax.set_ylim(-0.5, 5.)
    plt.xlabel('$\\tau$')
    plt.ylabel('$a,H,s$')
    plt.xscale('linear')
    ax.axvline(x[imin],color="grey", ls='dotted', zorder=-1)
    ax.axvline(1.,color="grey", ls='dotted', zorder=-1)
    ax.axhline(0.,color="grey", ls='dotted', zorder=-1)
    ax.axhline(1.,color="grey", ls='dotted', zorder=-1)
    plt.plot(x[imin:imax], y[0,imin:imax], label='a')
    plt.plot(x[imin:imax], y[1,imin:imax], label='H')
    plt.plot(x[imin:imax], y[2,imin:imax], label='s')
    plt.legend()
    plt.show()
    return
```

```python
#Calculate distance modulus
def calc_mu(AX, y0, y, yp, nz, zt, Hz, di, mu, *args):
    H0   = args[4]*2.13312e-42
    H0_1 = 1/H0
    nt = len(zt)
    
    # Solve Friedman equations
    imin, imax, x, y = calc_aHs(AX, y0, yp, *args)

    # Integrand of luminosity distance    
    a = y[0]
    H = y[1]
    for i in range(nt):
        az = 1/(zt[i]+1)
        az2 = az**2
        for j in range(imin+1,imax):
            if a[j] > az:
                Hz[i] = H[j]+(H[j]-H[j-1])*(az-a[j-1])/(a[j]-a[j-1])
                if i < nz: di[i] = H0_1/Hz[i]
                break

    # mu
    dz2 = 0.5*(z[1]-z[0])
    mu[0] = dz2*(di[1]+di[0])
    for i in range(1,nz):
        mu[i] = mu[i-1]+dz2*(di[i]+di[i-1])
    for i in range(nz):
        mu[i] = 5*np.log10((1+z[i])*mu[i]/1.5637382e38)+25
    return
```

```python
# S_k(r)
def S_k(Ok, H0C, r, prc):
    res = 0
    if abs(Ok) < prc:
        res = r
    elif Ok < 0:
        KHC = np.sqrt(-Ok)*H0C
        res = np.sin(KHC*r)/KHC
    else:
        KHC = np.sqrt(Ok)*H0C
        res = np.sinh(KHC*r)/KHC
    return res


# CMB distance priors, mu_s
def calc_R_la(nz, zt, Hz, rcd, mu_s, *args, prc=1.e-10):
    H0C  = args[4]*2.13312e-42
    Om = args[0]
    Ok = args[3]
    nt = len(zt)
    ztr = 1089  # transparency
    H0C2 = 0.5/H0C
   
    # rcd(z)
    rcd[0] = 0  # rcd(z*)
    ntr = 0
    for i in range(1,nt):
        dz2 = H0C2*(zt[i]-zt[i-1])
        rcd[i] = rcd[i-1]+dz2*(1/Hz[i]+1/Hz[i-1])
        if zt[i] > ztr: ntr = i-1
    rtr = rcd[ntr]
    print("rtr: ",rtr)
    
    # sound horizon rs
    rs  = 0
    igr0 = 1/(np.sqrt(1+660/(1+zt[ntr]))*Hz[ntr])
    for i in range(ntr+1,nt):
        dz2 = (zt[i]-zt[i-1])/2
        igr = 1/(np.sqrt(1+660/(1+zt[i]))*Hz[i])
        rs += dz2*(igr+igr0)
        igr0 = igr
    rs /= (H0C*np.sqrt(3))
    
    # R, la
    if Om < 0 or abs(rs) < prc:
        R = la = 0
        err = 1
    else:
        skr = S_k(Ok, H0C, rtr, prc)
        R  = np.sqrt(Om)*H0C*skr
        la = np.pi*skr/rs
        err = 0
        
    # mu_s(z)
    for i in range(1,nt):
        mu_s[i] = 5*np.log10((1+zt[i])*S_k(Ok, H0C, rcd[i], prc)/1.5637382e38)+25
    mu_s[0] = mu_s[1]
    return R,la,err
```

```python
GR_asym = 'n'
if GR_asym == 'n':
    #          Om,    Orad,    Ol,     Ok,    h,     Os      Og     s1
    args  = (0.315, 5.44e-5, 1.013,  0.000, 0.674, 2.700, -1.000, 0.500)
else:
    #          Om,    Orad,    Ol,     Ok,    h,     Os
    args1 = (0.315, 5.44e-5, 0.000,  0.000, 0.674, 0.900)
    Ol_GR = 0.685
    Og    = Ol_GR*(3-5*args1[5])/(1-args1[5])**2/2  # Og
    s1    = np.sqrt(2*Ol_GR/(1-args1[5]))          # s1
    args  = (args1+(Og,s1))
print("args: ",args)

AX, y0, y, yp, nz, z, Hz, di, mu, zt, rcd, mu_s = prepare_solving()
test_aHs(AX, y0, yp, *args)
print("y0: ",y0)
starttime = time.time()
calc_mu(AX, y0, y, yp, nz, zt, Hz, di, mu, *args)
elapsedtime = time.time()-starttime
print('\n elapsed time [sec] = ', elapsedtime)

R,la,err = calc_R_la(nz, zt, Hz, rcd, mu_s, *args)
print("R,la: ",R,la,err)

# plot H(z)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_xlim(zt[0], zt[(len(zt)-1)])
ax.set_ylim(-1.e1, 5.e1)
plt.xlabel('$z$')
plt.ylabel('$H(z)$')
ax.axhline(Hz[len(zt)-1],color="grey", ls='dotted', zorder=-1)
ax.plot(zt, Hz,'k')
plt.show()

# plot mu, mu_s, pantheon
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_xlim(0, 2)
ax.set_ylim(30, 50)
plt.errorbar(redshift, distmod, yerr=distmod_unc, capsize=3, fmt='r.', label='Pantheon data')
plt.plot(z, mu, label='$\mu$')
plt.plot(zt, mu_s, color='k',label='$\mu_s$')
plt.xlabel('$z$')
plt.ylabel('$\mu$')
plt.legend(loc='lower right')
plt.show()
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_xlim(zt[0], zt[(len(zt)-1)])
plt.errorbar(redshift, distmod, yerr=distmod_unc, capsize=3, fmt='r.', label='Pantheon data')
plt.plot(zt, mu_s, color='k',label='$\mu_s$')
plt.xlabel('$z$')
plt.ylabel('$\mu$')
plt.legend(loc='lower right')
plt.show()

# rcd
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_xlim(zt[0], zt[(len(zt)-1)])
plt.plot(zt, rcd, color='k',label='$r_{cd}$')
plt.xlabel('$z$')
plt.ylabel('$r_{cd}$')
plt.legend(loc='lower right')
plt.show()
```

```python
# Define the log_likelihood functions
def log_likelihood_SNe(params):
    Om, Ok, Os = params
    #Omn = ((Om-0.31)/0.01)
    args = (Om, Oradiation, 1.013,  Ok, 0.674, Os, -1.000, 0.500)
    #AX, y0, y, yp, nz, z, Hz, di, mu, zt, rcd, mu_s = prepare_solving()
    calc_mu(AX, y0, y, yp, nz, zt, Hz, di, mu, *args)
    f = interpolate.interp1d(z, mu, kind = 'quadratic')
    return -0.5*np.sum(((distmod.values - f(redshift))/distmod_unc.values)**2 + np.log(2*np.pi*distmod_unc**2))

#def log_likelihood_CC(params):
#    Om, Ok, Os = params
#    args = (Om, Oradiation, 1.013,  Ok, 0.674, Os, -1.000, 0.500)
#    _,_,_,y = calc_aHs(AX,y0,yp,*args) #y[1] is H, y[0] is a
#    z_st = (1/y[0]) - 1
#    f = interpolate.interp1d(z_st, y[1]*67.4, kind = 'linear') #check the *67.4
#    #x = np.linspace(0,2,1000)
#    #plt.figure()
#    #plt.plot(x,f(x))
#    #plt.show()
#    return -0.5*np.sum(((Hub - f(redshift_CC))/Hub_unc)**2 + np.log(2*np.pi*Hub_unc**2))

#def log_likelihood_BBN(params):
#    Om, Ok, Os = params
#    args = (Om, Oradiation, 1.013,  Ok, 0.674, Os, -1.000, 0.500)
#    _,_,_,y = calc_aHs(AX,y0,yp,*args) #y[1] is H, y[0] is a
#    z_st = (1/y[0]) - 1
#    f = interpolate.interp1d(z_st, y[1]*67.4, kind = 'linear') 
#    return -0.5*((HLCDM - f(10**9))/HLCDM)**2

#def log_likelihood_CMB(params):
#   Om, Ok, Os = params
#    args = (Om, Oradiation, 1.013,  Ok, 0.674, Os, -1.000, 0.500)
#........................
```

```python
# Uniform log_prior
def log_prior(params):
    Om, Ok, Os = params
    if 0.3 <= Om <= 0.4 and -0.01 <= Ok <= 0.01 and 1.0 <= Os <= 3.0:
        return -np.log(0.1*0.1*2)
    return -np.inf
```

```python
#Product, resp. sum in logarithm form of likelihood and prior
def log_sum(params):
    # Compute the log-prior
    lp = log_prior(params)
    if lp == -np.inf:
        # If the prior is zero (i.e. log is -infinity), return negative infinity
        return -np.inf
    
    # Compute the log-likelihood
    lSNe = log_likelihood_SNe(params)
    #lCC = log_likelihood_CC(params)
    #lBB = log_likelihood_BBN(params)
    #lCMB = log_likelihood_CMB(params)
    
    # Compute the log-posterior as the sum of the log-prior and log-likelihood
    return lp + lSNe
```

```python
# Set the number of dimensions (i.e. the number of parameters)
ndim = 3

# Set the number of walkers
nwalkers = 9

# Define the range of values for each parameter
theta_min = [0.3, -0.01, 1.0]
theta_max = [0.4, 0.01, 3.0]

# Generate random starting positions for the walkers within the defined range
p0 = np.zeros((nwalkers, ndim))
for i in range(ndim):
    p0[:, i] = np.random.uniform(theta_min[i], theta_max[i], size=nwalkers)

# Initialize the sampler
sampler = emcee.EnsembleSampler(nwalkers, ndim, log_sum, args=())
```

```python
# Run the sampler for a specified number of iterations
n_iterations = 1000
sampler.run_mcmc(p0, n_iterations, progress=True)
```

```python
# Get the chain of samples produced by the sampler
#samples = sampler.chain[:, :, :].reshape((-1, ndim))
samples = sampler.chain[:, :, :]
print(sampler.chain[:,:,:].shape)

# Discard the first few samples as burn-in
burnin = 100
final_samples = samples[:,burnin:, :]
print(final_samples.shape)
```
